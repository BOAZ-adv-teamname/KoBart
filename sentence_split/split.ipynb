```python
from kss import split_sentences,split_chunks
import pandas as pd
import konlpy
```

```python
df=pd.read_csv('../data/train.tsv',sep='\t')
text=df.iloc[12,1]
```

### library 자체 split
kss, hannanum 이용

```python
#kss
text1=split_sentences(text)
text1
```

```python
split_chunks(text,max_length=128)
```

```python
split_sentences(text1[1])
```

```python
from koalanlp.Util import initialize, finalize
from koalanlp import API
from koalanlp.proc import *
```

```python
initialize(java_options="-Xmx4g", HNN='2.1.4')
splitter = SentenceSplitter(api=API.HNN)
paragraph=splitter.sentences(text)
# 사용이 끝나면 반드시 finalize를 호출합니다
finalize()

paragraph
```

형태소를 통해 문장분리
형태소 분석기 + koalanlp => 문장분리

```python
initialize(java_options="-Xmx4g", HNN='2.1.4')

tagger = Tagger(API.HNN)

tagged_sentence = tagger.tagSentence(text)
paragraph = SentenceSplitter.sentencesTagged(tagged_sentence[0])
finalize()
paragraph
```

```python
initialize(java_options="-Xmx4g", KKMA='2.1.4')

tagger = Tagger(API.KKMA)

tagged_sentence = tagger.tagSentence(text)
paragraph = SentenceSplitter.sentencesTagged(tagged_sentence[0])
finalize()
paragraph
```
